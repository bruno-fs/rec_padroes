## Data preprocessing

Because there are missing values in the original dataset (`nan` values in `D_Matrix`), and the clustering algorithm might be affected by them, we need to either impute them or delete them. For that, we can do the following steps to preprocessing the data.

  1. Filter out those genes with more than 2 missing values (i.e., just consider genes with at most 2 missing values)
  2. Impute the missing values by using $k$-nearest neighbour (KNN) approach, where $k = 15$ (this value can be changed).
  3. Standardize the data, i.e., for each gene, make its mean 0 and its standard deviation 1.

### Missing values - KNNimpute algorithm

In order to proceed with Step 2, this paper [[2](http://www.ncbi.nlm.nih.gov/pubmed/11395428)] describes a way to impute missing values. 

The KNN-based method selects genes with expression profiles similar to the gene of interest to impute missing values. If we consider gene A that has one missing value in experiment 1, this method would find K other genes, which have a value present in experiment 1, with expression most similar to A in all other experiments (from 2 to N) (where N is the total number of experiments). A weighted average of values in experiment 1 from the K closest genes is then used as an estimate for the missing value in gene A. In the weighted average, the contribution of each gene is weighted by similarity of its expression to that of gene A.

After examining a number of metrics for gene similarity (Pearson correlation, Euclidean distance, variance minimization), we determined that Euclidean distance was a sufficiently accurate norm. This finding is somewhat surprising, given that the Euclidean distance measure is often sensitive to outliers, which could be present in microarray data. However, we found that log-transforming the data seems to sufficiently reduce the effect of outliers on gene similarity determination.


## Clusterização Hierárquica

Para fazer clusterização hierárquica, é necessário definir uma métrica("distância") entre pontos. Para este trabalho, vamos usar a mesma forma de calcular a distância utilizada no trabalho de Eise et al. [[3](http://www.pnas.org/content/95/25/14863.full)].

### Similaridade entre pontos

The gene similarity metric we use is a form of correlation coefficient. For any two genes $X$ and $Y$ observed over a series of $N$ conditions, a similarity score can be computed as follows:



$$S(X,Y)=\frac{1}{N}\sum_{i=1}^{N} \frac{X_i-\overline{X}}{\sigma_X}\frac{Y_i-\overline{Y}}{\sigma_Y}$$

where

$$\sigma_Z=\sqrt{\frac{1}{N}\sum_{i=1}^{N} (Z_i-\overline{Z})^2}$$

Na verdade, a distância entre $X$ e $Y$, denotada por $dist(X,Y)$ é:

$$dist(X,Y)=1-S(X,Y)$$

Esta distância é conhecida por *correlation distance*.

### Cálculo da matriz de distância

Usando esta forma de calcular a "distância" entre todos os pontos, dois a dois. Desta forma, podemos montar uma matriz de distâncias `DM` (*distance matrix*), onde `DM[i][j]` é a distância do gene `i` ao gene `j`, representados pelos vetores numéricos `D_Matrix[i]` e `D_Matrix[j]`.

Esta matriz de distâncias pode ser obtida usando a função `distance` do módulo `scipy.spatial` como mostrada abaixo:

```python
import scipy.spatial.distance as distance
pairwise_dists = distance.squareform(distance.pdist(DD_Matrix,metric='correlation'))
pairwise_dists_columns = distance.squareform(distance.pdist(DD_Matrix.T,metric='correlation'))
```

Desta forma, `pairwise_dists[i][j]` é a distância dos  numéricos `D_Matrix[i]` e `D_Matrix[j]`.

## Clusterização Hierárquica

The method `sch.linkage` from `scipy.cluster.hierarchy` module performs hierarchical clustering. As you would expect, you can specify the method of clustering with the method parameter and the metric with the metric. 

Para isso, você deverá colocar no início de seu programa python:

```python
import scipy.cluster.hierarchy as sch
```

Relembrando que para clusterização hierárquica, existem pelo menos 4 métodos para unir grupos: `complete`, `single`, `average`, e `centroid`. Para este trabalho, vamos utilizar o `average`.

```python
# cluster
clusters = sch.linkage(pairwise_dists, method='average')
```

Em um `ipython notebook` (veja neste [link](http://nbviewer.ipython.org/github/ucsd-scientific-python/user-group/blob/master/presentations/20131016/hierarchical_clustering_heatmaps_gridspec.ipynb)), encontra-se a seguinte explicação para o método `linkage` do `scipy.cluster.hierarchy`:

`sch.linkage` returns a matrix with 4 columns and $n-1$ rows. Each row records which two clusters were combined as the heirarchical clustering was performed. For instance, if the first row is

    [  5.        ,   8.        ,  13.91203052,   2.        ],

then clusters 5 and 8 were combined because they had a distance of 13.91203052. The 2 means that there are two original observations in the newly formed cluster. If the second row is 

    [  3.           9.          17.77496623   2.        ],

then clusters 3 and 9 were combined because they had a distance of 17.8 and there are two original observations in the new cluster etc.

Aliás, neste `ipython notebook` ([link](http://nbviewer.ipython.org/github/ucsd-scientific-python/user-group/blob/master/presentations/20131016/hierarchical_clustering_heatmaps_gridspec.ipynb)), explica como fazer um `dendrograma` usando o método `sch.dendrogram`. Sugiro fortemente você olhar este `ipython notebook` para gerar um dendrograma de seu agrupamento.

Por exemplo, o código abaixo plota um dendrograma dos 4185 genes agrupados usando `distance correlation` e `average linkage`.

```python
import matplotlib.gridspec as gridspec

# helper for cleaning up axes by removing ticks, tick labels, frame, etc.
def clean_axis(ax):
    """Remove ticks, tick labels, and frame from axis"""
    ax.get_xaxis().set_ticks([])
    ax.get_yaxis().set_ticks([])
    for sp in ax.spines.values():
        sp.set_visible(False)

testDF = pd.DataFrame(np.array(DD_Matrix, dtype='float'))
# print testDF.shape

fig = plt.figure(figsize=(10, 100))
heatmapGS = gridspec.GridSpec(1,2,wspace=0.0,hspace=0.0,width_ratios=[0.25,1])

### row dendrogram ###
denAX = fig.add_subplot(heatmapGS[0,0])
denD = sch.dendrogram(clusters,orientation='right')
clean_axis(denAX)

### heatmap ###
heatmapAX = fig.add_subplot(heatmapGS[0,1])
axi = heatmapAX.imshow(testDF.ix[denD['leaves']],interpolation='nearest',aspect='auto',origin='lower',cmap=plt.cm.RdBu)
clean_axis(heatmapAX)
```
